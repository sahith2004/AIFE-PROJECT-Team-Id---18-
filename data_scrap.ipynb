#STOCK MARKET DATA SCRAPPING 

TOP 5 NSE STOCKS DOWNLOAD DATA


```python
from NSEDownload import stocks
from datetime import datetime
import pandas as pd

stock_symbols = ["SBIN", "HDFC", "TATASTEEL", "ICICIBANK", "TATAMOTORS"]



dfs = []

for symbol in stock_symbols:
    try:
        df = stocks.get_data(stock_symbol=symbol, start_date='1-1-2013', end_date='1-1-2023')

        dfs.append(df)

   

if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
    print("Combined data for all stocks:")
    print(combined_df)
else:
    print("No data available for analysis")
```

```python
import os

data_folder = 'data'
os.makedirs(data_folder, exist_ok=True)

csv_file_path = os.path.join(data_folder, 'stock_data.csv')

combined_df.to_csv(csv_file_path, index=False)

print(f"Data saved to {csv_file_path}")
```
